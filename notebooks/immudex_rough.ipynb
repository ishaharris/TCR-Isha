### Write a function to extract hits


```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Set correct count matrix and CMV labels
os.chdir("/Users/ishaharris/Projects/TCR/TCR-Isha/data/Count_Matrices")
count_matrix = 'IM2_countmatrix_ng.tsv'

cmv_labels = ['CMV', 'CMV 2', 'CMV_68_R1', 'CMV_31_R1', 'CMV_68_R2', 'CMV_31_P2', 'CMV_31', 'CMV_A1101_31']
```

```python

def extract_matches(cmv_labels, count_matrix):
    # Set wd
    os.chdir("/Users/ishaharris/Projects/TCR/TCR-Isha/data/Count_Matrices")
    # Read the count matrix
    df = pd.read_csv(count_matrix, sep="\t")
    
    # Clean NAs
    df = df.dropna(subset=['clonotype_id'])

    # Filter for CMV>10
    valid_cmv_labels = [label for label in cmv_labels if label in df.columns]
    df_filtered = df.loc[df[valid_cmv_labels].gt(10).any(axis=1)]

    # Print how many unique clonotypes are in the filtered dataframe
    print(f"{df_filtered['clonotype_id'].nunique()} unique clonotypes in {count_matrix}")

    # Save CMV-associated clonotypes and their counts 
    export_df=df_filtered[['clonotype_id','cdr3s_aa']+valid_cmv_labels].reset_index(drop=True)

    #Write the clonotype_counts to a CSV file
    os.chdir("/Users/ishaharris/Projects/TCR/TCR-Isha/data/High_confidence_CMV_TCR")
    output_file = f"{count_matrix.split('_')[0]}_CMV_counts.csv"
    export_df.to_csv(output_file, index=False, header=True)
    print(f"CMV-associated clonotype counts saved to {output_file}")
    
    return export_df


```

```python
count_matrix = 'IM3_countmatrix_ng.tsv'
extract_matches(cmv_labels, count_matrix)
```

```python
count_matrix = 'IM3_countmatrix_ng.tsv'
extract_matches(cmv_labels, count_matrix)
```

```python
# Visualise clonotype size distribution
def clonotype_size_distribution(count_matrix):
    # Read the count matrix
    df = pd.read_csv(count_matrix, sep="\t")
    
    # Clean NAs
    df = df.dropna(subset=['clonotype_id'])
    
    # Get clonotype counts
    clonotype_counts = df['clonotype_id'].value_counts()

    # Sample name
    sample_name = count_matrix.split('_')[0]
    
    # Plot
    clonotype_counts.head(20).plot(kind="bar")
    plt.xlabel('Clonotype ID')
    plt.ylabel('Number of Cells')
    plt.title(f'{sample_name} Clonotype Size Distribution')
    plt.tight_layout()
    plt.show()
```

### Compare all 3 lists of hits and curate into 1 master

```python
import re
# Set wd

os.chdir("/Users/ishaharris/Projects/TCR/TCR-Isha/data/High_Confidence_CMV_TCR")
#read pd
im1_hits = pd.read_csv("IM1_CMV_counts.csv", sep=",")
im2_hits = pd.read_csv("IM2_CMV_counts.csv", sep=",")
im3_hits = pd.read_csv("IM3_CMV_counts.csv", sep=",")

# Concatenate the dataframes
cmv_hits = pd.concat([im1_hits, im2_hits, im3_hits], ignore_index=True)

# Add a new column for just TRB cdr3s
cmv_hits['TRB'] =cmv_hits['cdr3s_aa'].str.extract(r'TRB:([^;]+)') 

# Count number of unique TRB
unique_trb = cmv_hits['TRB'].nunique()
```

```python
import pandas as pd

# Extract the number following either 'IM' or 'IMD' at the start of the string
cmv_hits['IM_group'] = cmv_hits['clonotype_id'].str.extract(r'^IMD?(\d)').squeeze().apply(lambda x: f"IM{x}")

# Reorder
cols = cmv_hits.columns.tolist()
cols.insert(1, cols.pop(cols.index('IM_group')))
cmv_hits = cmv_hits[cols]
```

```python
# Export CMV hits
os.chdir("/Users/ishaharris/Projects/TCR/TCR-Isha/data/High_confidence_CMV_TCR")
cmv_hits.to_csv("CMV_hits.csv", index=False, header=True)
```
