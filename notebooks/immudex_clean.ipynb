```python
import os
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Load in count matrix using path
path = Path("/Users/ishaharris/Projects/TCR/TCR-Isha/data/Count_Matrices")
cm = pd.read_csv(path / f, sep="\t")
```

```python
rearr = pd.read_csv(path / "AC02_rearrangement.tsv", sep="\t")
```

```python
rearr_clean = rearr[['cell_id', 'clone_id', 'v_call', 'd_call','j_call','c_call', 'junction_aa']].copy()
```

```python
# reduce cm to one row/barcode
cm_pos = (
    cm[['Barcode','positive']]
      .drop_duplicates(subset='Barcode')
)

# now merge—you’ll get exactly one positive per rearr row
merged = (
    rearr_clean
      .merge(cm_pos,
             left_on='cell_id',
             right_on='Barcode',
             how='left')
      .drop(columns='Barcode') 
)

merged
```

```python
# Save merged to count_matrices

merged.to_csv(path / "MERGED_AC02_countmatrix.tsv", sep="\t", index=False)
```

### Identify EBV cross reactors

```python
# get list of columns
cols = cm.columns.tolist()
print(cols)
# get list of peptide_ cols
peptide_cols = [col for col in cols if col.startswith('peptide_')]
peptide_cols

# filter for any peptide > 10
ebv = cm[cm[peptide_cols].gt(10).any(axis=1)]

# remove ebv cross reactors from cmv
ebv_cmv = ebv[ebv['positive'] > 10]

# remove na in clonotype_id
ebv_cmv = ebv_cmv[~ebv_cmv['clonotype_id'].isna()]

# count rows where clonotype_id, v_call and j_call are unique
ebv_cmv_unique = ebv_cmv.drop_duplicates(subset=['clonotype_id', 'v_call', 'j_call'])

# reset indices
ebv_cmv_unique = ebv_cmv_unique.reset_index(drop=True)
ebv_cmv_unique

# drop rows where v call starts with 'TRA'
ebv_cmv_unique = ebv_cmv_unique[~ebv_cmv_unique['v_call'].str.startswith('TRA')]
ebv_cmv_unique

```

```python
# Split rows for every unique TRB
df = ebv_cmv_unique.copy() # reset this as desired

def split_trb_rows(row):
    trb_parts = [part for part in row['cdr3s_aa'].split(';') if part.startswith('TRB:')]
    rows = []
    for trb in trb_parts:
        new_row = row.copy()
        new_row['cdr3s_aa'] = trb.replace('TRB:', '')  # remove 'TRB:' prefix  # keep only this TRB string
        rows.append(new_row)
    return rows

expanded_rows = [split_trb_rows(row) for _, row in df.iterrows()]
flattened_rows = [row for sublist in expanded_rows for row in sublist]

# Create a new DataFrame from the flattened rows
expanded_df = pd.DataFrame(flattened_rows)
# Reset the index
expanded_df.reset_index(drop=True, inplace=True)

ebv_cmv_unique = expanded_df
```
